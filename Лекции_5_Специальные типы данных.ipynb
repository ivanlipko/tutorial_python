## Cписки - lists
Список - это своего рода контейнер, куда можно положить, что душе угодно. Элементом списка может быть как число, так и строка и даже другой список.
Для создания списка используют квадратные скобки, либо функцию list():

```python
a = 'миру мир'

spisok = [2021, "СевГУ", a.split(' ')]
print(spisok)

# можно создавать пустые списки
spisok = []

spisok = list('СПИСОК')  # функция list принимает 1 аргумент, который состоит из элементов
                         # такие объекты называются итерируемыми, строки и списки - итерируемые объекты
print(spisok)
```

Списки упорядочены, то есть порядок элементов имеет значение. Поэтому, как и строки, списки поддерживают индексирование. Все операции по выделению элементов и срезов аналогичны таковым со строками.

В отличии от строк, списки - объекты изменяемые. То есть можно поменять один элемент списка на другой, добавить или удалить элементы.

```python
print(spisok[2::2])

spisok[0] = "В"
print(spisok)

# для добавления элементов в конец списка используется метод .append(элемент)
spisok.append('spam')

# метод .pop(индекс) возвращает элемент списка с заданным индексом, а сам элемент из списка удаляет
# если индекс не указать, то .pop() применится к последнему элементу списка
# для удаления елемента по значению используется метод .remove(значение), который удалит первый
# попавшийся элемент с таким значением
poped = spisok.pop(1)
spisok.remove('И')
spisok.remove('В')
print('spisok = ', spisok, '\npoped = ', poped)

# как и любой итерируемый объект, список имеет длину, которую можно получить с помощью len(список)
print('Длина spisok: ', len(spisok))

# также можно посчитать число элементов, имеющих определенное значение
print('Число вхождений "spam" в spisok: ', spisok.count('spam'))

# cписки, как и строки, поддерживают конкатенацию
spisok += [a] * 2 + [b, c]

# список можно отсортировать..
spisok.sort()
print(spisok)
# .. и развернуть
spisok.reverse()
print(spisok)
```

### Передача списка в функцию

Демонстрируется функция, которая получает список в качестве параметра. Затем в теле функции этот список изменяется (изменяются первый и второй элементы списка).

Список - изменяемый объект


```python
# Функция, которая получает параметром список
def ChangeList(L):
    # Изменить список
    L[0] = 777 # Изменить первый элемент списка
    L[1] = (2, 3) # Изменить второй элемент списка

    # Вывести список в теле функции
    print('ChangeList.L = ', L)

# Создать список
LL = [ [2, 3.5], True, 'abcd' ]

# Вывести список LL для контроля
print('LL = ', LL)

# Вызвать функцию и передать ей список LL
ChangeList(LL)

# Вывести список LL после вызова функции
print('LL = ', LL)
```

Изменение списка в теле функции приводит к изменению этого списка в вызывающем коде. 

При передаче списка в функцию можно использовать операцию среза `[:]`, как показано в следующей программе. Срез `LL[:]` осуществляет копию списка. Таким образом, изменения списка внутри функции не приведут к изменению списка в вызывающем коде, поскольку эти изменения будут выполняться над копией.

```python
LL = [ [2, 3.5], True, 'abcd' ]

# Вызвать функцию и передать ей список как срез [:]
ChangeList(LL[:])

# Вывести список L после вызова функции
print('LL = ', LL)
```

### Кортежи - tuples
Кортеж - это всего-лишь неизменяемый список.
Любые операции со списками, не изменяющие его, применимы и к кортежам.

```python
# создается кортеж аналогично списку, но только круглыми списками или функцией tuple()
kortej = (1, 2, 5, 3, 1)
print('Число единиц в кортеже: ', kortej.count(1))
```

Зачем они нужны?
- Защита от дурака
- Занимают меньше места в памяти
- Бывают ситуации, где просто нельзя использовать изменяемый объект

```python
tuple_size = kortej.__sizeof__()
list_size = list(kortej).__sizeof__()

print(
    'Размер кортежа: ', tuple_size, 'байт(а)',
    '\nРазмер списка: ', list_size, 'байт(а)'
)
```

#### Передача кортежа в функцию

```python
def PassTuple(T):
    # Изменить кортеж
    T = ([1,2], 'a', 25.88)
    print('PassTuple.T = ', T)

# Создать кортеж
TT = ( [2, 3.5], True, 'abcd')

# Вызвать функцию и передать ей кортеж TT
PassTuple(TT)

# Вывести кортеж TT
print('TT = ', TT)
```

Передача кортежа в функцию происходит по значению. Содержимое не изменяется.

### Словари - dicts
Кортежи и списки - это очень полезные структуры, однако, иногда поиск элементов по индексу неудобен. Ведь индекс ничего нам не говорит о том, что именно за элемент за ним скрывается.
Если элементы большие и сложные или если мы хотим создать понятную иерархическую структуру, на помощь нам придут словари.

Представьте, что вы в библиотеке. Разве в поисках нужной книги вам бы хотелось копаться в огромной горе из всех книг?
Куда проще копаться в стопке листиков с названиями книг, на которых написано, где каждую конкретную книгу можно найти.

В словарях в роли названий выступают ключи - keys. Ключи, конечно, должны быть неизменяемыми объектами (строки, кортежи). Представьте, что вы пришли в библиотеку, но пока вы шли, название искомой книги изменилось - это будет печально.

Каждый ключ связан со значением. В роли значения может быть любая структура, в том числе список, другой словарь и т.д.

```python
# словарь создается с помощью фигурных скобок или функцией dict(пары ключ-значение),
# где аргумент - это итерируемый объект, элементы которого - это пары ключ - значение
slovarik = dict( [
        ('a', 1),
        ('b', 2),
        ('c', 3)
    ] )

# или, что то же самое
slovarik = dict(
    a = 1,
    b = 2,
    c = 3
)
print(slovarik)

slovarik2 = {
    'Учитель': 'Марья Петровна',
    'Ученики': ('Маша', 'Петя')
}
print(slovarik2)
```

Словарь - объект, вообще говоря, неупорядоченный. Поэтому индексирование он не поддерживает.
Зато поддерживает обращение по ключу:

```python
print(slovarik2['Учитель'])

# Можно создать новый ключ, присваивая значение по несуществующему ключу
slovarik2['Директор'] = 'Сан Саныч'

# Значения по существующему ключу можно изменять (но не сами ключи)
slovarik2['Учитель'] = slovarik2['Директор']

print(slovarik2)

# Можно получить значение по ключу, удалив сам ключ из словаря
director = slovarik2.pop('Директор')
print('Извлекли: ', director, '\nСловарь: ', slovarik2)

keys = slovarik2.keys()  # возвращает список ключей словаря
values = slovarik2.values()  # возвращает список значений словаря
items = slovarik2.items()  # возвращает список пар ключ-значение

print(
    'Ключи: ', keys,
    '\nЗначения: ', values,
    '\nПары ключ-значение: ', items
)
```

```python
2.7  Перебор элементов словаря

Можно легко организовать перебор ключей всех элементов в словаре:

Capitals = {'Russia': 'Moscow', 'Ukraine': 'Kiev', 'USCapitals': 'Washington'}
for key in Capitals:
    print(key, Capitals[key])

Вспомните о методах keys, values и items.

Быстро проверить, если ли значение val среди всех значений элементов словаря Capitals можно так: val in Capitals.values(), а организовать цикл так, чтобы в переменной key был ключ элемента, а в переменной val было его значение можно так:

Capitals = {'Russia': 'Moscow', 'Ukraine': 'Kiev', 'USCapitals': 'Washington'}
for key, val in Capitals.items():
    print(key, val)

```

#### Передача словаря в функцию

Демонстрируется функция ChangeDict() которая получает словарь в качестве параметра. В теле функции изменяются отдельные элементы словаря, лежащие на позициях с номерами 0 и 2.

Словарь - изменяемый объект


```python
# Функция, получающая параметром словарь
def ChangeDict(D):
    # Изменить словарь
    D['A'] = 100 # Изменить первый элемент списка
    D['C'] = 300 # Изменить третий элемент списка

    # Вывести список в теле функции
    print('ChangeDict.D = ', D)

# Создать словарь
DD = { 'A':5, 'B':10, 'C':20 }

# Вывести словарь DD для контроля
print('DD = ', DD)

# Вызвать функцию и передать ей словарь DD
ChangeDict(DD)

# Вывести список DD после вызова функции
print('DD = ', DD)
```

**Сдвинуть список**

Чтобы по месту изменить список, можно срез использовать:

```python
def spin_forward(array):
    array[:] = array[-1:] + array[:-1]
    
arr = [1,2,3,4,5,6,7]
spin_forward(arr)
arr
```

Или можно вернуть новое значение, не трогая входной список:

```python
def spinned_forward(array):
    return array[-1:] + array[:-1]

arr = [1,2,3,4,5,6,7]
out = spinned_forward(arr)
print(out)
```
