### Синтаксис, Синтаксические правила

Документ `PEP8` определяет "соглашения" по написанию кода. Соглашения помогают всем писать более понятные и читабельные программы.

* В идентификаторах (имена переменных, функций, типы и т.д.) учитывается регистр.

    Пример разных переменных:
       
|||
|:-|:-:|
|A    a||
|name  Name  NAME| |
|cat  Cat  CAT| |


Правила именования 
    
* Именовать переменные со смыслом
        some_usefull_variable
        more_than_other_value
        is_sunny_day
        y и x если другого смысла нет?

* Нет двойному использованию переменных. Например, если переменную `name` вы используете для хранения имени, то не надо её использовать для хранения фамилии.
* Идентификатор должен начинаться с буквы (от a до Z) или со знака подчеркивания (_), после которых может идти произвольное количество букв, знаков подчеркивания и чисел (от 0 до 9). 
* @ # % и др. специальные символы нельзя использовать в качестве идентификаторов
* имена переменных начинаются с маленькой буквы, имена классов с большой. Можно использовать `CamelStyle` или `snake_case`

* Использования знака подчеркивания в качестве первого символа идентификатора означает, что данный идентификатор является частным (закрытым от использования вне класса).
* Если идентификатор начинается и заканчивается двумя знаками подчеркивания (например, __init__ ) это означает, что он является специальным именем, определенным внутри языка.

## Запуск интерпретатора Python

В настоящее время существует две версии языка Python: более старая, но пока ещё более распространённая версия 2 и современная версия 3. Мы будем использовать версию 3 данного языка. Именно её необходимо установить дома, скачав данную версию с сайта www.python.org.

Программирование может выполняться в двух режимах: интерактивный, когда вы пишите код и сразу его выполняете или "программируемый", когда вы сначала пишите целиком программу или её блоки, а только потом выполняете.


### Интерактивный режим


Запустить интерпретатор python можно из командной строки:

Ёлочка `>>>` обозначет что это строка ввода, приглашение командной строки чтобы ввести команду. Строка без этой ёлочки содержит результат, который вывел интерпретатор.

Вводите команды и наслаждайтесь результатом. А что можно вводить? Несколько примеров:


Первая команда вычисляет сумму двух чисел, вторая команда вычисляет 2 в степени 100, третья команда выполняет операцию конкатенации для строк, а четвертая команда печатает строку 'ABC', повторенную 10 раз.

Выход из интерпретатора выполняется по команде `exit()`. Именно со скобочками, так как это — функция или с помощью комбинации `Ctrl+D`.

### Программируемый режим

Интерактивный режим подходит для простых разовых команд. Когда необходимо создавать более сложные программы, то их пишут в текстовом редакторе, там же отлаживает и проверяют. Теперь создадим программу и выполним её целиком:

````python
a = 358
b = 197
c = (a ** 2 + b ** 2) ** 0.5
print(c)
````

Здесь мы используем переменные -- объекты, в которых можно сохранять различные (числовые, строковые и прочие) значения. В первой строке переменной `a` присваивается значение 358, затем переменной `b` присваивается значение 971, затем переменной `c` присваивается значение арифметического выражения, равного длине гипотенузы. После этого значение переменной `c` выводится на экран с помощью команды `print`.

При этом значения вычисленных выражений не выводятся на экран (в отличии от интерактивного режима), поэтому для того, чтобы вывести результат работы программы, то есть значение переменной `c`, нужна функция `print()`.

## Основные арифметические операторы
Операция присваивания "=" присваивает переменной, находящейся слева от оператора, значение, справа от него. 

* в Питоне "=" это на самом деле не присвоение, а связывание

````python
a = 2
print(a)
````

В отношении чисел операции сложения, вычитания, умножения и деления обычно не вызывают в понимании. В Питоне имеются операторы инкремента, декремента и схожие для домножения и деления.

````python
foo = 1 + 2
foo += 1
print(foo)

foo -= foo - 1
print(foo)

foo *= 2
foo /= 3
print(foo)
````

Существуют операции возведения в степень и целочисленного деления:

````python
a = 2
b = a**2
print(b)
b **= 2
print(b)

print( 'Целочисленное деление', (b + 1) // a, '\n',
    'Остаток от деления', (b + 1) % a  )
````

Полезно помнить, что квадратный корень из числа `x` — это `x**0.5`, а корень степени `n` — это `x**(1/n)`.

## Общие правила определения приоритетов операций:

* Выполняются возведения в *степень справа налево*, например,  `3**3**3` это $3^{27}$.
* Выполняются унарные минусы (отрицания).
* Выполняются умножения и деления *слева направо*. Операции умножения и деления имеют одинаковый приоритет.
* Выполняются сложения и вычитания *слева направо*. Операции сложения и вычитания имеют одинаковый приоритет.
* Операции сравнения `>  <  ==`
* Операция присваивания `=`

Питон позволяет присваивать значения сразу нескольким переменным.

## Конвертация числовых типов

В Питоне существует такой механиз типизации как "утиная типизация" (Ducktyping). Например,  если `А` ведёт себя как утка (крякает, плавает и проч), то она является уткой. Отсюда нет особой строгости к работе с переменными. 

> Но это накладывает требование на то, чтобы сам программист следил за типами данных.

Например, числа и строки поддерживают операцию "+". Но для них эта операция работает по разному:

````python
n1, n2 = 10, 5
num = n1 + n2
print(type(n1), type(n2), type(num))
print(num)

text = 'let '+ 'me ' + 'know'
print(type(text), text)

text = 'let me know' + num
print(type(text))
````

Целые int 

Дробные float (с плавающей точкой).

Тип данных можно узнать, применив стандартную функцию type.
Стоит сказать о конвертации данных одного типа в другой, об округлении и о делении.

````python
a = 4 / 2
print(type(a))
print(a)

a = int(a)  # динамическая типизация
print(type(a))
print(a)

b = 0.2 + .3 + 2e-01 + a  # автоматическое приведение к одному типу
print(type(b))
print(b)
 
a = int(b)  # приведение к целому числу включает в себя округление в МЕНЬШУЮ сторону
print(type(a))
print(a)

a = round(b)  # используйте функцию round для округления в ближайшую сторону
print(type(a))
print(a)
````

Нужно помнить, что вычисления с использованием чисел с плавающей точкой имеют погрешность! В основном это касается операций умножения и деления, особенно, если они проводятся между очень большим и очень маленьким числами.

## Комментарии

В бывают ситуации ,когда в тексте программы нужно "скрыть" от исполнения часть кода, но не удалять, написать поясняющий текст. Для этого служат комментарии. 

````python
# Однострочный комментарий 

''' Это НЕ Много
    строчный
    комментарий.
    Это строка!!!
'''
````

Строковые литералы, заключенные в тройные кавычки, могут содержать:

* кавычки `"`
* апострофы `'`
* *docstring*-комментарии `"""func desc"""`
* переводы строк 

В тоже время, стоит помнить, что такой строковый литерал не должен содержать внутри символов `'''`. 

Руководство по стилю кода (pep-8) рекомендует использовать # для блочных комментариев. 

Примеры:

````python
x = 5
# x = 6
x = x + 2
print(x)
````
